```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../csvs/placement-dataset-binary.csv")
# dataset.drop(["Unnamed"], inplace=True)
dataset
```

```python
dataset.drop(columns=[dataset.columns[0]], inplace=True)
dataset
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x="cgpa", y="iq", data=dataset, hue="placement")
```

```python
x = dataset.iloc[:,:-1]
x.ndim
# checks dimension of data here 2
```

```python
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=28)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
x
```

```python
lr.predict(X=[x.iloc[4]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```
