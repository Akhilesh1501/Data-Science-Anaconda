```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../csvs/Social_Network_Ads.csv")
dataset
```

```python
dataset.drop(columns=["EstimatedSalary", "Gender", "User ID"], inplace=True)
```

```python
dataset
```

```python
sns.scatterplot(x="Age", y="Purchased", data=dataset)
```

```python
X = dataset[["Age"]]
Y = dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[58]])
```

```python
sns.scatterplot(x="Age", y="Purchased", data=dataset)
sns.lineplot(x="Age", y = lr.predict(X), data=dataset , c="red")
```
