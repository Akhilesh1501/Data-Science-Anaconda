```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../csvs/bank-loan.csv")

dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset["othdebt"].fillna(dataset["othdebt"].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# sns.distplot(dataset["othdebt"])
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
ss.fit(dataset[["othdebt"]])
```

```python
ss.transform(dataset[["othdebt"]])
```

```python
dataset["othdebt_ss"] =  pd.DataFrame(ss.transform(dataset[["othdebt"]]), columns=["x"])
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["othdebt_ss"])

plt.subplot(1,2,2)
plt.title("After")

sns.distplot(dataset["othdebt"])
plt.show()
```
