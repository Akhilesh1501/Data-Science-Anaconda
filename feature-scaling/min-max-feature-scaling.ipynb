```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../csvs/bank-loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["othdebt"].fillna(dataset["othdebt"].mean(), inplace=True)
dataset.isnull().sum()
```

```python
sns.distplot(dataset["othdebt"])
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
ms.fit(dataset[["othdebt"]])
```

```python
ms.transform(dataset[["othdebt"]])
```

```python
dataset["othdebt_ms"] = pd.DataFrame(ms.transform(dataset[["othdebt"]]))
```

```python
dataset.describe()
```

```python
plt.figure(figsize= [10,5])
plt.subplot(2,3,1)
plt.title("Before")
sns.distplot(dataset["othdebt"])


plt.subplot(2,3,2)
plt.title("After")
sns.distplot(dataset["othdebt_ms"])

plt.subplot(2,3,3)
plt.title("After")
sns.distplot(dataset["othdebt_ms"])
```
